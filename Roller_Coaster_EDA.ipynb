```python
df.rename(columns={'coaster_name':'Coaster_name',
       'year_introduced':'Year_introduced', 'latitude':'Latitude', 'longitude':'Longitude', 'Type_Main':'Type_main',
       'opening_date_clean':'Opening_date_clean', 'speed_mph': 'Speed_mph', 'height_ft':'Height_ft'},inplace = True)
```

```python
df['Opening_date_clean'].dtype
```

Here the dtype `<M8[ns]` refers to as follows 
- " < " refers to little-Endian  = low significance values first
- *M8* refers to datetime datatype
- *[ns]* converys that precision is in nanoseconds

# Answering questions with data


```python
#What are the locations with high speed coasters and have minimum of 10 coasters 
df.columns
```

```python
df_ = df[df['Location']!='Other']\
      .groupby(['Location'])\
      ['Speed_mph']\
      .agg(['mean','count'])\
      .dropna().sort_values(by='mean',ascending=True).query('count>=10')
df_.plot(kind='barh',y='mean',\
         title='Locations with high speed coasters and minimum of 10 coasters',\
         figsize=(12,5),color='b')
```
